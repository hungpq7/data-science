## 4.5. Formatting

### Basic formatting
The <code style='font-size:13px;'>str.format()</code> method fills its arguments to the placeholder (identified with the braces <code style='font-size:13px;'>{}</code>):
- If a placeholder is empty then it will be filled with the corresponding argument
- If a placeholder contains a number, that number acts as the index of the filling argument
- If a placeholder contains a name, it will be filled with the value of that variable

````python
'The answer is {}'.format(3+4)
````

````python
'{} and {} and {}'.format('a', 1, None)
````

````python
'{1} {0} {0} {1}'.format('three', 'seven')
````

````python
'The {cat} is {color}'.format(cat='panther', color='black')
````

## 4.6. Regular expression
A regular expression (regex) is a sequence of characters that defines a search pattern. For example, <code style='font-size:13px;'>^a...s$</code> defines any five-letter string starting with <code style='font-size:13px;'>a</code> and ending with <code style='font-size:13px;'>s</code>. Python has the module <code style='font-size:13px;'>re</code> to work with regex.

### Search patterns
Metacharacters are characters that are interpreted in a special way by a regex engine. The metacharacters are:
<code style='font-size:13px;'>.</code>
<code style='font-size:13px;'>^</code>
<code style='font-size:13px;'>$</code>
<code style='font-size:13px;'>*</code>
<code style='font-size:13px;'>+</code>
<code style='font-size:13px;'>?</code>
<code style='font-size:13px;'>[]</code>
<code style='font-size:13px;'>()</code>
<code style='font-size:13px;'>{}</code>
<code style='font-size:13px;'>\\</code>
<code style='font-size:13px;'>|</code>.

````python
import re
````

A period <code style='font-size:13px;'>.</code> matches any single character, except for <code style='font-size:13px;'>\\n</code>. The <code style='font-size:13px;'>re.findall()</code> function finds all the substrings that match the search pattern.

````python
# return all 3-character substrings
re.findall('...', 'abcd\n1234')
````

````python
# return all 3-character substrings, including "\n"
re.findall('...', 'abcd\n1234', flags=re.DOTALL)
````

The caret <code style='font-size:13px;'>^</code> is used to check if the string starts with certain characters. The dollar sign <code style='font-size:13px;'>$</code> is used to check if the string ends with certain characters.

````python
# return the first 3 characters if the original string starts with "ab"
re.findall('^ab.', 'abcdefgf')
````

````python
# return the last 4 characters if the original string ends with "89"
re.findall('..89$', '0123456789')
````

Adding <code style='font-size:13px;'>flags=re.MULTILINE</code> will search the pattern in every line instead of the entire string only, and return as much text as possible.

````python
string = '''
abcd
amnd
abcde
aabcdd
'''
re.findall('^a..d$', string, flags=re.MULTILINE)
````

The following metacharacters match a number of occurrences of the preceding character:  
- The asterisk <code style='font-size:13px;'>*</code> matches 0 or more repetitions.
- The plus sign <code style='font-size:13px;'>+</code> matches 1 or more repetitions.
- The question mark <code style='font-size:13px;'>?</code> matches 0 or 1 repetitions.
- The braces <code style='font-size:13px;'>{x,y}</code> match at least x and at most y repetitions.

For example:
- <code style='font-size:13px;'>ma*n</code> matches
<code style='font-size:13px;'>mn</code>
<code style='font-size:13px;'>man</code>
<code style='font-size:13px;'>maan</code>
<code style='font-size:13px;'>maaan</code>
- <code style='font-size:13px;'>ma+n</code> matches
<code style='font-size:13px;'>man</code>
<code style='font-size:13px;'>maan</code>
<code style='font-size:13px;'>maaan</code>
- <code style='font-size:13px;'>ma?n</code> matches
<code style='font-size:13px;'>mn</code> and <code style='font-size:13px;'>man</code> only
- <code style='font-size:13px;'>ma{2,4}n</code> matches <code style='font-size:13px;'>maan</code> <code style='font-size:13px;'>maaan</code> and <code style='font-size:13px;'>maaaan</code> only

````python
re.findall('ma+n', 'mn man maan maaan main')
````

````python
re.findall('ma{2,3}n', 'mn man maan maaan main')
````

The brackets <code style='font-size:13px;'>[]</code> specify a set of character need to match. The hyphen <code style='font-size:13px;'>-</code> can be used to specify a range of characters and the caret <code style='font-size:13px;'>^</code> can be used to denote that the characters not in the set will be match.

````python
re.findall('[Aa]', 'Anaconda')
````

````python
re.findall('[a-e]', 'jupyter notebook')
````

````python
re.findall('[^1-9]', 'abc123')
````

The backslash <code style='font-size:13px;'>\\</code> behaves as an escape character.
- <code style='font-size:13px;'>\\d</code> matches any digit. Equivalent to <code style='font-size:13px;'>[0-9]</code>. The opposite: <code style='font-size:13px;'>\\D</code>.
- <code style='font-size:13px;'>\\s</code> matches if a string contains any whitespace character such as space <code style='font-size:13px;'> </code>, new line <code style='font-size:13px;'>\\n</code> and tab <code style='font-size:13px;'>\\t</code>. The opposite: <code style='font-size:13px;'>\\S</code>.
- <code style='font-size:13px;'>\\w</code> matches any alphanumeric character (digits and alphabets). Equivalent to <code style='font-size:13px;'>[a-zA-Z0-9]</code>. The opposite: <code style='font-size:13px;'>\\W</code>.

````python
re.findall('\d', 'a1b2c34')
````

````python
re.findall('\s', 'a string that spans\nmultiple lines')
````

### Regex functions

````python
import re
````

The `re.findall` function returns a list of substrings that matches the search pattern.

````python
re.findall('\d+', 'date:08month:12year:2019')
````

The `re.split` function splits the original string using search pattern as the separators.

````python
re.split(':\d+', 'date:08month:12year:2019')
````

````python
re.split(':\d+', 'date:08month:12year:2019', maxsplit=2)
````

The <code style='font-size:13px;'>re.sub()</code> function replaces all matches with the content of the <code style='font-size:13px;'>repl</code> parameter. The <code style='font-size:13px;'>re.subn()</code> function works the same, but it returns both the replaced string and the number of substitutions made.

````python
re.sub(
    pattern=r'\s+',
    repl='_',
    string='tiger  lion cheetah \n leopard \t jaguar'
)
````

````python
re.sub(
    pattern=r'\s+',
    repl='_',
    string='tiger  lion cheetah \n leopard \t jaguar',
    count=2
)
````

````python
re.subn(
    pattern=r'\s+',
    repl='_',
    string='tiger  lion cheetah \n leopard \t jaguar'
)
````

The <code style='font-size:13px;'>re.search()</code> and <code style='font-size:13px;'>re.match()</code> functions both return an <code style='font-size:13px;'>re.match</code> object.

````python
if re.match('...', 'abcdef'):
    print(True)
else:
    print(False)
````

````python
# return 2 subgroups
match = re.search(r'(\d{3}) (\d{2})', '39801 356, 2102 1111')
print(match.group())
print(match.group(1))
print(match.groups())
````

````python
# return the indices of the first and the last character of the match
match = re.search(r'\d+', 'abc12345de')
print(match.start())
print(match.end())
print(match.span())
````

## 5.3. Datetime formatting
Using the
<code style='font-size:13px;'>dt.datetime.strftime()</code>
and
<code style='font-size:13px;'>dt.datetime.strptime()</code>
methods. *strftime* stands for *string-format-time* and *strptime* stands for *string-parse-time*.

### Date and time format codes
Code|Meaning                         |Example                             |
:---|:-------------------------------|:-----------------------------------|
`%y`|Year, 2 last digits             |00, 01, 02,..., 99                  |
`%Y`|Year, full                      |0001, 0002,..., 2019, 2020,..., 9999|
`%b`|Month name, abbreviated         |Jan, Feb,..., Dec                   |
`%B`|Month name, full                |January, February,..., December     |
`%m`|Month of the year, zero-padded  |01, 02, 03,..., 12                  |
`%W`|Week of the year, zero-padded   |00, 01, 02,..., 53                  |
`%d`|Day of the month, zero-padded   |01, 02, 03,..., 31                  |
`%a`|Weekday name, abbreviated       |Mon, Tue,..., Sun                   |
`%A`|Weekday name, full              |Monday, Tuesday,..., Sunday         |
`%H`|Hour, 24-hour clock, zero-padded|00, 01, 02,..., 23                  |
`%I`|Hour, 12-hour clock, zero-padded|01, 02, 03,..., 12                  |
`%p`|AM or PM                        |AM, PM                              |
`%M`|Minute, zero-padded             |00, 01, 02,..., 59                  |
`%S`|Second, zero-padded             |00, 01, 02,..., 59                  |

````python
import datetime as dt
````

````python
# strptime() converts a string to a datetime object
string_1 = '11/8/2018 12:18:00 PM'
datetime_1 = dt.datetime.strptime(string_1, '%m/%d/%Y %I:%M:%S %p')
datetime_1
````

````python
# strftime() converts a datetime object to a string
datetime_2 = dt.datetime(2018, 11, 8, 12, 18)
string_2 = datetime_2.strftime('%m/%d/%Y %I:%M:%S %p')
string_2
````

````python
# get date, day, month of the current day, print as "DD-MM-YYYY"
date = dt.date.today()
date.strftime('%d-%m-%Y')
````
